Este código utiliza o método de clessificação kNN para descobrir se um paciente que fez cirurgia de câncer de mama morreu ou sobreviveu, dadas entradas de idade, ano de operação, e número de nódulos encontrados. A saída indica: 1 sobreviveu 5 anos ou mais; 2 - Morreu dentro de 5 anos

Neste exemplo são utilizados 3 algoritmos de kNN - Proprio, exemplo do curso e utilizando sklearn

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
%%time
#Treinamento simples decorando os dados de treinamento. Dados dos indices pares do array de amostras
def train_kNN(dataset):
    return dataset[::2]
    
#efetua a distancia euclidiana entre os elementos 0 e 2 do array
def dist(el1, el2):
    s = 0.0
    for i in range(0,3,2):
        s += (el1[i] - el2[i])**2
    return np.sqrt(s)

def knn_impl2(treinamento, nova_amostra, K):
    dists, tam_treino = {}, len(treinamento)
    # calcula a distância euclidiana da nova amostra para
    # todos os outros exemplos do conjunto de treinamento
    for i in range(tam_treino):
        d = dist(treinamento[i], nova_amostra)
        dists[i] = d
    # obtém as chaves (índices) dos k-vizinhos mais próximos
    k_vizinhos = sorted(dists, key=dists.get)[:K]
    
    # votação majoritária
    qtd_rotulo1, qtd_rotulo2 = 0, 0
    for indice in k_vizinhos:
        if treinamento[indice][-1] == 1:
            qtd_rotulo1 += 1
        else:
            qtd_rotulo2 += 1
            
    if qtd_rotulo1 > qtd_rotulo2:
        return 1
    else:
        return 2

def sklearn_knn(train, test, k):
    knn = KNeighborsClassifier(n_neighbors=k)
    entradas , saidas, entradas_test = [],[],[]
    for t in train:
        entradas.append([t[0],t[2]])
        saidas.append(t[3])
    
    for t in test:
        entradas_test.append([t[0],t[2]])
    
    knn.fit(entradas,saidas)
    
    labels = knn.predict(entradas_test)
    
    acerto = 0 
    
    for i in range(len(entradas_test)):
        if labels[i] == test[i][-1]:
            acerto += 1
    acc = 100*(acerto/len(entradas_test))
    
    print("Validation accuracy for k=%d is %f%%" % (k,acc))
    
    
def classif_kNN(train, obj, k):
    dists = [] #Array para manter as distancias calculadas
    index = [] #Array para manter a referencia para o indice do dado referente a  distancia
    
    #Efetua a distancia do valor recebido (obj) com todos os outros dados
    #do conjunto de treinamento
    for i in range(len(train)):
        d = dist(train[i], obj)
        
        #adiciona elementos ate a quantidade de k elementos
        if len(dists) < k:
            dists.append(d)
            index.append(i)
        else:
            #faz a insercao ordenada dos elementos quando o array ja esta cheio
            jaux = 0
            daux = d
            for j in range(k):
                
                if dists[j] > daux:
                    daux = dists[j]
                    jaux = j
            if daux != d:
                index[jaux] = i
                dists[jaux] = d
                
    cl1 = 0 #classe 1 - viveu
    cl2 = 0 #classe 2 - Morreu
    
    #Faz a votacao majoritaria dos elementos
    for i in range(k):
        #print("Value for index " + str(index[i]) + " is " + str(train[index[i]][3]))
        if train[index[i]][3] == 1:
            cl1 += 1
        else:
            cl2 += 1

    if cl1 > cl2:
        return 1
    else:
        return 2

def validade_test_set(test_set, train, k):
    right = 0
    
    for i in range(len(test_set)):
        #predic = knn_impl2(train,test_set[i],k)
        predic = classif_kNN(train, test_set[i], k)
        
        if predic == test_set[i][3]:
            right +=1
            
    return (right/len(test_set))*100
        
def info_dataset(dataset, verbose=True):
    class1 = 0; class2 = 0
    
    for data in dataset:
        if data[-1] == 1:
            class1 +=1
        elif data[-1] == 2:
            class2 +=1
            
    if verbose:
        print("Class 1: %d" % class1)
        print("Class 2: %d" % class2)
        print("Total: %d" % len(dataset))
    return len(dataset),class1,class2


    
raw_data = np.genfromtxt('../../data/habermam.data', delimiter=',')
train = train_kNN(raw_data)
test = raw_data[1::2]

info_dataset(train)
k = 5


#Roda e valida a classificacao knn com sci-kit learn
sklearn_knn(train,test,k)

#Roda e valida os metodos pessale exemplo da aula (alterar o tipo na funcao validade_test_set)
#acc = validade_test_set(test, train, k)
#print("Validation accuracy for k=%d is %f%%" % (k,acc))

#print("Test input: ", test[10])
#print("Test output: ", classif_kNN(train, test[10], 13))

```
